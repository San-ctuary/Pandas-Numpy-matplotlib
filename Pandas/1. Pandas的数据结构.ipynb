## DataFrame

> DataFrame是一个表格型的数据结构，既有行索引也有列所以，可以被看做由Series组成的字典

传入一个等长列表或Numpy数组组成字典

````python
data = {'state':['Ohio','Ohio','Ohio','Nevada','Nevada'],
       'year':[2000,2001,2002,2001,2002],
       'pop':[1.5,1.7,3.6,2.4,2.9]}
````

````python
frame = DataFrame(data)
````

````python
frame #所有列按序排列
````

````python
#按照指定顺序排列
DataFrame(data,columns=['year','state','pop'])
````

````python
#传入的列找不到数据，输出NAN
frame2 = DataFrame(data,columns=['year','state','pop','debt'],index=['one','two','three','four','five'])
````

````python
frame2
````

````python
frame2.columns
````

````python
# 通过类似字典取值的方式将DataFrame的列获取为一个Series
frame2['state']
````

````python
# 取行可以用索引字段ix
frame2.ix['three']
````

````python
frame2.loc['three']
````

````python
# 列通过赋值的方式修改
frame2['debt'] = 16
````

````python
frame2
````

````python
frame2['debt'] = np.arange(5)
````

````python
frame2
````

````python
# 将Series赋值给某个列时，会精确匹配DataFrame的索引，空位会被填上缺失值
val = Series([-1.2,-1.5,-1.6],['one','two','three'])
````

````python
frame2['debt'] = val
````

````python
frame2
````

````python
# 给不存在的列赋值，会创建一个新列
frame2['eastern'] = frame2.state == 'Ohio'
````

````python
frame2
````

````python
#del 可以删除列
del frame2['eastern']
````

````python
frame2.columns
````

**注意：**对索引方式返回的列都是  视图，所以对返回的Series所做的修改都会反映到源DataFrame上

````python
temp_series = frame2['debt']
````

````python
frame2
````

````python
temp_series
````

````python
temp_series['one'] = 10000
````

````python
temp_series
````

````python
frame2  # 显然修改了源frame2的值
````
