```python
import pandas as pd
import numpy as np
```

```python
boolean = [True,False]
gender = ['男','女']
color = ["white","black","yellow"]
data = pd.DataFrame({
    "height": np.random.randint(150,190,100),
    "weight": np.random.randint(40,90,100),
    "smoker": [boolean[x] for x in np.random.randint(0,2,100)],
    "gender": [gender[x] for x in np.random.randint(0,2,100)],
    "age": np.random.randint(15,90,100),
    "color": [color[x] for x in np.random.randint(0,3,100)]
})
```

```python
data
```

# Map函数

## 将gender用0，1代替男女

```python
# map方法会把对应的数据逐个当作参数传入到字典或者函数
# 用字典映射 
data["gender"].map({"男":1,"女":0})
```

```python
# 定义函数
def gender_map(x):
    gender = 1 if x == "男" else 0
    return gender
data["gender"].map(gender_map)
```

```python
print(data["age"])
data["age"].apply(lambda x : x + 3)
```

# Apply
## Apply  处理多列数据效果较好

```python
data[["height","weight","age"]].apply(np.sum,axis=0)
```

```python
def BMI(series):
    weight = series["weight"]
    height = series["height"]/100
    BMI = weight / height ** 2
    return BMI

data["BMI"] = data.apply(BMI,axis=1)
data
```

## applymap
对单元格中所有的元素进行操作

```python
df = pd.DataFrame(
    {
        "A":np.random.randn(5),
        "B":np.random.randn(5),
        "C":np.random.randn(5),
        "D":np.random.randn(5),
        "E":np.random.randn(5),
    }
)
df
```

```python
df.applymap(lambda x : "%.2f" % x)
```
